```python
#LEGB Rule (Name Resolution Order)
#Local, Enclosing, Global, Builtins

x=10

def outer():
    #x="enclosing"
    def inner():
        #x="local"]
        print(x)
    inner()
outer()
```

```python
# python decides scope at compile time, not runtime
# this will give error because to python x is local everywhere in inner
# If a name is assigned anywhere in a function, 
# Python treats it as local everywhere in that function unless declared otherwise.
x="global"

def outer():
    #x="enclosing"
    def inner():
        print(x)
        x="local"
    inner()
outer()
```

```python
#Global and nonlocal scope modifiers

x=10
def f():
    #will tell to no create local binding and use the global one
    global x
    x=20
f()
x
```

```python
def outer():
    x=10
    def inner():
        nonlocal x
        x+=1
    inner()
    return x
outer()
```

```python
#defaults are evaluated once

def f(x, lst=[]):
    lst.append(x)
    return lst
f(1)
f(2)

#Correct way

def f1(x, lst=None):
    if lst is None:
        lst=[]
    lst.append(x)
    return lst
f1(1)
f1(2)
```
