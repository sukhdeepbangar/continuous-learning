# Iteration, iterables, iterators, and generators

A protocol is a set of method names that, if present on an object, cause Python to treat that object in a particular way.

Iterator Protocol

If object has 
__iter__()
__next__()
Then python allow
 for x in obj:
    ...

```python
# when python sees for x in obj:

# it translates into

_iter = iter(obj)
while True:
    try:
        x=next(_iter)
    except StopIteration:
        break
    ....
```

Iterable is an object with implements __iter__()
Returns an iterator

Iterator is an object which implements __next__()
and raise StopIteration when exhausted

yield suspends execution, can resume later


Generators are functions that return iterators

Generator is a state machine and single use by design

Generation Expressions can be used for lazy loading in Comprehensions for infinite sequences and streaming data

```python
def gen():
    yield 1
    yield 2
    yield 3 

next(gen())
next(gen())
```
