# Python's data model and attribute access


What obj.attr really means

it does not directly read memory.
it executes type(obj).__getattribute__(obj,"attr")
Inside that 
    Check obj.__dict__
    Check class dict
    Check base classes

If not found
    Call __getattr__ if defined

Else 
    Raise Attribute Error

```python
class A:
    def __getattr__(self, name):
        return f"Missing: {name}"

a=A()
a.x
```

```python
class A:
    x=10
    #This intercepts all attribute accesses, should not overrride
    def __getattribute__(self, name):
        return f"Intercepted"
a=A()
a.x
```

```python
class A:
    def f (self):
        return "Hello"
    
a=A()
A.__dict__['f'].__get__(a)()

```

Functions implements desriptor protocol
A descriptor is any object with one or more of:
"__get__"
"__set__"
"__delete__"
"property" is a descriptor
Functions implement
function.__get__(instance, owner)

so 
a.f is actually
A.f.__get__(a,A)

Operation Overloading is method dispatch

a + b is a.__add__(b)

same for

== -> __eq__
< -> __lt__
[] -> __getitem__
len() -> __len__
